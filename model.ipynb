```python
import cv2
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from keras.preprocessing import image
import sys
from keras.models import Sequential, load_model, model_from_json
from keras.models import Sequential, load_model, model_from_json
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
classes = {
    0: 'fist',
    1: 'ok',
    2: 'palm',
    3: 'peace',
    4:'swing'
}
gesture_model = load_model('gesture_predict_v4.h5', compile=False)
```

```python
cam = cv2.VideoCapture(0)

ok, frame = cam.read()
bg = frame.copy()

kernel = np.ones((3,3),np.uint8)
if not cam.isOpened():
    print("Could not open video")
    sys.exit()
positions = {
    'hand_pose': (65, 100)}
cv2.namedWindow("Capture")
while True:
    # Read a new frame
    ret, frame = cam.read()
    diff = cv2.absdiff(bg, frame)
    mask = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    # Threshold the mask
    th, thresh = cv2.threshold(mask, 20 ,255,cv2.THRESH_BINARY)
    # Opening, closing and dilation
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
    img_dilation = cv2.dilate(closing, kernel, iterations=2)

    frame_predict = np.expand_dims(cv2.resize(img_dilation, (160, 213)), axis=0).reshape((1, 160, 213, 1))
    frame_predict = frame_predict/255
    try:
        prediction = gesture_model.predict(frame_predict)
        print(prediction)
        predi = prediction[0].argmax()
        gesture = classes[predi]
        for i, pred in enumerate(prediction[0]):
                # Draw confidence bar for each gesture
                barx = positions['hand_pose'][0] 
                bary = 60 + i*60
                bar_height = 20
                bar_length = int(400 * pred) + barx # calculate length of confidence bar

                # Make the most confidence prediction green
                if i == predi:
                    colour = (0, 255, 0)
                else:
                    colour = (0, 0, 255)

                cv2.putText(frame, "{}: {}".format(classes[i], pred), (positions['hand_pose'][0], 30 + i*60), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (255, 255, 255), 2)
                cv2.rectangle(frame, (barx, bary), (bar_length, bary - bar_height), colour, -1, 1)

        cv2.putText(frame, "hand pose: {}".format(gesture), positions['hand_pose'], cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0, 0, 255), 2)
        
    except Exception as ex:
        cv2.putText(frame, "hand pose: error", positions['hand_pose'], cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0, 0, 255), 2)
        cv2.putText(frame, "hand pose: error", positions['hand_pose'], cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0, 0, 255), 2)
   
    cv2.imshow("Threshold", thresh)
    cv2.imshow("img_dilation", img_dilation)
    cv2.imshow("Camera", frame)
    k = cv2.waitKey(1) & 0xff

    if k == 27: break # ESC pressed
cv2.destroyAllWindows()
cam.release()
```
